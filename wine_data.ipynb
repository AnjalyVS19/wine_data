```python
import matplotlib.pyplot as plt
plt.scatter(df[0],df[13])
plt.xlabel('type of wine')
plt.ylabel('quality')
plt.title('wine against quality')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x=df.iloc[:,0:13]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=32)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
k=model.predict(xtest)
k
```

```python
model.coef_
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
for i,pred in enumerate(k):
    print('Predicted value',pred,'Actual value',ytest.iloc[i])
```
